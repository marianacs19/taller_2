```python
def anios_educacion(df, nivel_col='P6210', grado_col='P6210s1',
                    contar_preescolar=False, nan_to_zero=False):
    import numpy as np, pandas as pd
    map_nivel = {'Ninguno':1,'Preescolar':2,'Básica primaria':3,'Basica primaria':3,'Primaria':3,
                 'Básica secundaria':4,'Basica secundaria':4,'Secundaria':4,
                 'Media':5,'Superior o universitaria':6,'Superior':6,'Universitaria':6,
                 'No sabe, no informa':9}
    niv = df[nivel_col]
    if niv.dtype == 'O': niv = niv.map(map_nivel)
    else: niv = pd.to_numeric(niv, errors='coerce')

    g = pd.to_numeric(df[grado_col], errors='coerce')
    y = pd.Series(np.nan, index=df.index)

    y.loc[niv==1] = 0
    y.loc[niv==2] = 1 if contar_preescolar else 0
    y.loc[niv==3] = g.clip(0,5)
    y.loc[niv==4] = g.clip(6,9)
    y.loc[niv==5] = g.clip(10,13)
    cond_sup = (niv==6) & g.notna()
    y.loc[cond_sup & (g<=8)] = 11 + g.loc[cond_sup & (g<=8)]
    y.loc[cond_sup & (g>8)]  = g.loc[cond_sup & (g>8)]
    y.loc[niv==9] = np.nan

    if nan_to_zero:
        y = y.fillna(0).astype('Int64')
    else:
        y = y.astype('Float64')
    return y
```

```python
def recode_one_to_zero_else_one(s: pd.Series) -> pd.Series:
    """
    Recode a series so that:
      - exact 1 -> 0
      - any other value (e.g., 0, 2, 9, 98, 99, negatives, strings) -> 1
      - NaN stays NaN (change behavior below if you prefer)
    """
    out = pd.Series(np.where(s.eq(1), 0, 1), index=s.index, name=s.name)
    return out

def recode_1_2_else_nan(s: pd.Series, return_nullable_int: bool = True) -> pd.Series:
    """
    Recode values as:
      1 -> 1
      2 and anything else -> 0
    """
    out = pd.to_numeric(s, errors="coerce").map({1: 1, 2: 0, 9: 0, 99: 0, 98: 0, np.nan: 0})
    if return_nullable_int:
        # Use pandas' nullable integer dtype so NaN becomes <NA> (not float NaN)
        out = out.astype("Int64")
    return out

def recode_cols_1_2_else_nan(df: pd.DataFrame, cols) -> pd.DataFrame:
    df_1 = df.copy()
    for c in cols:
        df_1[c] = recode_1_2_else_nan(df_1[c])
    return df_1


import pandas as pd

def broadcast_from_leader(
    df: pd.DataFrame,
    hh_col: str,          # household id, e.g., "id_house"
    leader_col: str,      # column indicating leader, e.g., "is_leader" (1=yes, 0=no)
    attrs: list[str],     # columns to broadcast from the leader
    leader_value=1,
    suffix="_hh"
) -> pd.DataFrame:
    out = df.copy()

    # keep only leader rows per household (if multiple leaders, take first non-null per attr)
    leaders = (
        out.loc[out[leader_col] == leader_value, [hh_col] + attrs]
        .sort_values([hh_col])  # optional, but deterministic
        .groupby(hh_col, as_index=False)
        .first()
    )

    # map each attribute back to all rows in that household
    for col in attrs:
        mapping = dict(zip(leaders[hh_col], leaders[col]))
        out[col + suffix] = out[hh_col].map(mapping)

    return out
```

```python
# Mapeos CIUO-08 (gran grupo -> nombre y nivel de competencias)
# consistente con https://www.dane.gov.co/files/sen/nomenclatura/ciuo/CIUO_08_AC_2015_07_21.pdf
CIUO08_MAJOR_NAMES = {
    0: "Fuerzas militares",
    1: "Directores y gerentes",
    2: "Profesionales, científicos e intelectuales",
    3: "Técnicos y profesionales del nivel medio",
    4: "Personal de apoyo administrativo",
    5: "Trabajadores de los servicios y vendedores de comercios y mercados",
    6: "Agricultores y trabajadores calificados agropecuarios, forestales y pesqueros",
    7: "Oficiales, operarios, artesanos y oficios relacionados",
    8: "Operadores de instalaciones y máquinas y ensambladores",
    9: "Ocupaciones elementales",
}

# Según CIUO-08: 0 → niveles 1–4; 1 → 3–4; 2 → 4; 3 → 3; 4–8 → 2; 9 → 1
CIUO08_SKILL_LEVELS = {
    0: 0, # acá sería 1,2,3,4 pero lo diferenciamos porque los militares tienen diferentes competecnias 
    1: 5, # me lo inventé porque en el CIUO 8 ponen 3,4 y acá suponemos que tienen un nivel especial de competencias
    2: 4,
    3: 3,
    4: 2,
    5: 2,
    6: 2,
    7: 2,
    8: 2,
    9: 1,
}

def add_ciuo08_major_group(
    df: pd.DataFrame,
    code_col: str = "Valor",
    out_code_col: str = "ciuo08_gran_grupo",
    out_name_col: str = "ciuo08_gran_grupo_nombre",
    out_skill_col: str = "ciuo08_nivel_competencias",
) -> pd.DataFrame:
    """
    Clasifica códigos CIUO-08 de 2+ dígitos en los 10 grandes grupos usando el primer dígito.
    - Maneja enteros, strings y NaN.
    - Si el código no es numérico o está fuera de 0–99, deja NA.
    """
    df1 = df.copy()
    
    # Asegura numérico; strings no convertibles → NaN
    codes = pd.to_numeric(df1[code_col], errors="coerce")

    # Gran grupo = primer dígito (entero // 10)
    major = (codes // 10).astype("Int64")

    # Filtra fuera de rango (negativos, >=10) a NA
    major = major.where(major.between(0, 9), pd.NA)

    # Crea columnas
    df1[out_code_col] = major
    df1[out_name_col] = major.map(CIUO08_MAJOR_NAMES)
    df1[out_skill_col] = major.map(CIUO08_SKILL_LEVELS)

    return df1


# Definir función para asignar pesos según edad
def asignar_peso(edad):
    if 24 <= edad <= 54:
        return 0.5# más peso para "prime age"
    elif edad < 24:
        return 0.2   # menos peso para jóvenes
    else:
        return 0.3   # un poco menos para mayores de 54
# Calcular promedio ponderado por hogar
def promedio_ponderado(grupo):
    return np.average(grupo['Edad'], weights=grupo['peso'])

# Función robusta para promedio ponderado por hogar
def wavg(grp):
    m = grp['Edad'].notna() & grp['peso'].notna()
    if m.any() and grp.loc[m, 'Edad'].sum() > 0:
        return np.average(grp.loc[m, 'Edad'], weights=grp.loc[m, 'peso'])
    return np.nan
```

```python
import pandas as pd

def collapse_constants_all_groups(df, id_col="id"):
    # Identify columns that are constant within each id for ALL groups
    const_mask = df.groupby(id_col).nunique(dropna=False).eq(1).all()
    const_cols = const_mask[const_mask].index.tolist()

    # Build the collapsed df (one row per id)
    out = (df.groupby(id_col, as_index=False)[const_cols]
             .first())  # .first() is fine because they’re constant
    return out

# usage
train_collpase_hogar = collapse_constants_all_groups(train, "id")
test_collpase_hogar = collapse_constants_all_groups(test, "id")
```

```python
import numpy as np
import pandas as pd
from itertools import combinations
from pandas.api.types import is_numeric_dtype


def impute_by_grid(
    df: pd.DataFrame,
    target_col: str,
    feature_cols: list,
    bin_specs: dict | None = None,
    default_bins: int = 3,
    default_method: str = "quantile",  # "quantile" (qcut) o "uniform" (cut)
    inplace: bool = False,
    keep_bin_cols: bool = False,
):
    """
    Imputa valores faltantes en 'target_col' usando promedios por celdas de una grilla
    definida a partir de 'feature_cols' (que NO deben tener NaNs).

    Parámetros
    ----------
    df : DataFrame de entrada.
    target_col : str
        Columna objetivo que tiene NaNs a imputar (numérica).
    feature_cols : list[str]
        Columnas sin NaNs que definen la grilla (se binarizan/segmentan).
    bin_specs : dict | None
        Especificación de bins por columna. Claves = nombres de columnas en feature_cols.
        Cada valor puede ser:
          - lista/array de cortes (ej.: [0, 20, 80, np.inf])
          - int (número de bins; usa cuantiles por defecto)
          - str con forma "quantile:K" o "uniform:K"
        Si None o si una columna no está en el dict, se usan `default_method` y `default_bins`.
    default_bins : int
        Número de bins por defecto si no se especifica para una columna.
    default_method : {"quantile","uniform"}
        Método por defecto para crear bins cuando no hay especificación.
    inplace : bool
        Si True, modifica df en el sitio. Si False, devuelve una copia.
    keep_bin_cols : bool
        Si True, mantiene las columnas *_bin con los intervalos.

    Retorna
    -------
    DataFrame con 'target_col' imputada (si inplace=False) o None (si inplace=True).

    Ejemplo rápido
    --------------
    # Bins explícitos para edad: 0-20, 20-80, 80+
    bin_specs = {"edad": [0, 20, 80, np.inf]}
    out = impute_by_grid(df, "ingreso", ["edad", "estrato"], bin_specs=bin_specs)
    """

    if not inplace:
        df = df.copy()

    # Validaciones básicas
    if target_col not in df.columns:
        raise KeyError(f"'{target_col}' no está en el DataFrame.")
    missing_feat = [c for c in feature_cols if c not in df.columns]
    if missing_feat:
        raise KeyError(f"Columnas de grilla no encontradas: {missing_feat}")
    if not is_numeric_dtype(df[target_col]):
        df[target_col] = pd.to_numeric(df[target_col], errors="coerce")
    # Garantía (opcional): feature_cols sin NaNs (el usuario lo prometió, pero por si acaso)
    for c in feature_cols:
        if df[c].isna().any():
            raise ValueError(f"La columna '{c}' tiene NaNs; la grilla exige columnas sin NaNs.")

    bin_specs = bin_specs or {}

    def _make_bins(series: pd.Series, spec):
        """Devuelve la serie binned (Interval) a partir de la especificación."""
        if spec is None:
            # Método por defecto
            if default_method == "quantile":
                # qcut intenta equilibrar conteos por bin
                return pd.qcut(series, q=default_bins, duplicates="drop")
            elif default_method == "uniform":
                return pd.cut(series, bins=default_bins, include_lowest=True, right=False)
            else:
                raise ValueError("default_method debe ser 'quantile' o 'uniform'.")

        # Int -> cuantiles por defecto
        if isinstance(spec, int):
            return pd.qcut(series, q=spec, duplicates="drop")

        # String "quantile:K" o "uniform:K"
        if isinstance(spec, str):
            try:
                method, k = spec.split(":")
                k = int(k)
            except Exception:
                raise ValueError(f"Especificación inválida para bins: {spec!r}")
            method = method.strip().lower()
            if method == "quantile":
                return pd.qcut(series, q=k, duplicates="drop")
            elif method == "uniform":
                return pd.cut(series, bins=k, include_lowest=True, right=False)
            else:
                raise ValueError(f"Método no soportado en spec: {method}")

        # Lista/array de cortes explícitos
        try:
            edges = np.asarray(spec, dtype=float)
            if np.isneginf(edges[0]):
                # Por seguridad, cambiar -inf por mínimo real
                edges[0] = series.min()
            return pd.cut(series, bins=edges, include_lowest=True, right=False)
        except Exception as e:
            raise ValueError(f"No se pudo interpretar 'bin_specs' para la columna: {series.name}") from e

    # Crear columnas binned
    bin_cols = []
    for c in feature_cols:
        spec = bin_specs.get(c, None)
        b = _make_bins(df[c], spec)
        bcol = f"{c}__bin"
        df[bcol] = b
        bin_cols.append(bcol)

    # Promedio por la grilla completa
    # Nota: transform ignora NaNs del target al promediar
    group_cols = bin_cols
    fill_series = df.groupby(group_cols, dropna=False)[target_col].transform("mean")

    # Backoff jerárquico: si una celda no tiene promedio (todo NaN),
    # probamos con promedios en grupos más agregados (todas las combinaciones)
    if fill_series.isna().any():
        # De mayor granularidad a menor
        for k in range(len(group_cols) - 1, 0, -1):
            for subset in combinations(group_cols, k):
                candidate = df.groupby(list(subset), dropna=False)[target_col].transform("mean")
                fill_series = fill_series.fillna(candidate)
            if not fill_series.isna().any():
                break

    # Promedio global como último recurso
    if fill_series.isna().any():
        global_mean = df[target_col].mean()
        fill_series = fill_series.fillna(global_mean)

    # Imputar solo en filas con NaN en target
    mask_nan = df[target_col].isna()
    df.loc[mask_nan, target_col] = fill_series[mask_nan].values

    if not keep_bin_cols:
        df.drop(columns=bin_cols, inplace=True)

    return None if inplace else df
```

```python
# crea 'resultado' con la cascada central -> segunda -> tercera
train_collpase_hogar["Arriendo_sumado"] = train_collpase_hogar[["Arriendo_pagado_mensual", "Arriendo_estimado_mensual", "Pago_cuota_amortizacion_mensual"]].bfill(axis=1).iloc[:, 0]
test_collpase_hogar["Arriendo_sumado"] = test_collpase_hogar[["Arriendo_pagado_mensual", "Arriendo_estimado_mensual", "Pago_cuota_amortizacion_mensual"]].bfill(axis=1).iloc[:, 0]
```
